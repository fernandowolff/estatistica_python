```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Analisando dados TMDB

```python
tmdb = pd.read_csv('./data/tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
ax = sns.displot(tmdb.vote_average, kind='kde')
ax.set_axis_labels("Nota média", "Densidade")
```

```python
ax = sns.displot(tmdb.vote_average, kind='hist')
ax.set_axis_labels("Nota média", "Frequência")
plt.title("Nota média dos filmes IMDB 5000")
```

```python
sns.boxplot(x = tmdb.vote_average).set(xlabel = "Nota média do filme", title = "Distribuição de nota média dos filmes do TMDB 5000")
```

Muitos filmes mostrando média 0, o que é estranho, pois dentre multiplas avaliações tais filmes ainda terem média 0, é bem improvável.

```python
tmdb.query("vote_average == 0").head()
```

Pode-se observar que alguns filmes tiveram poucos, ou até mesmo nenhum, votos.
Tais filmes não devem fazer parte da análise.

```python
tmbd_com_mais_de_10_votos = tmdb.query("vote_count >= 10")
tmbd_com_mais_de_10_votos.describe()
```

```python
tmdb.query("vote_average==10")
```

```python
ax = sns.displot(tmbd_com_mais_de_10_votos.vote_average, kind='hist')
ax.set_axis_labels("Nota média", "Frequência")
plt.title("Nota média dos filmes IMDB 5000 dentre os filmes com 10+ votos")
```

```python
ax = sns.displot(tmbd_com_mais_de_10_votos.vote_average, kind='kde')
ax.set_axis_labels("Nota média", "Densidade")
plt.title("Média de votos em filmes no TMDB 5000 dentre os filmes com 10+ votos")
```

```python
ax = sns.displot(tmbd_com_mais_de_10_votos.vote_average, kind='ecdf')
ax.set_axis_labels("Nota média", "% acumulada de filmes")
plt.title("Média de votos em filmes TMDB com pelo menos 10 votos")
```

```python
sns.boxplot(x = tmbd_com_mais_de_10_votos.vote_average).set(xlabel = "Nota média do filme", title = "Distribuição de nota média dos filmes do TMDB 5000 dentre os filmes com 10+ votos")
```

# Analisando dados Movielens

```python
notas = pd.read_csv("./data/ratings.csv")
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
nota_media_por_filme.head()
```

```python
ax = sns.displot(nota_media_por_filme, kind='hist')
ax.set_axis_labels("Nota média", "Frequência")
plt.title("Nota média dos filmes MovieLens")
```

```python
quantidade_de_votos_por_filmes = notas.groupby("movieId").count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filmes.query("rating >= 10").index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.displot(nota_media_dos_filmes_com_pelo_menos_10_votos, kind='hist')
ax.set_axis_labels("Nota média", "Frequência")
plt.title("Nota média dos filmes MovieLens com pelo menos 10 votos")
```

```python
sns.boxplot(x = nota_media_dos_filmes_com_pelo_menos_10_votos).set(xlabel = "Nota média do filme", title = "Distribuição de nota média dos filmes do MovieLens dentre os filmes com 10+ votos")
```

Note que por mais que os datasets do TMDB e do MovieLens sejam diferentes, com valores mínimos e máximos diferentes, os graficos gerados quando se tratando de filmes com 10 ou mais votos, são, visualmente, bem similares.

```python
ax = sns.displot(nota_media_dos_filmes_com_pelo_menos_10_votos, kind='ecdf')
ax.set_axis_labels("Nota média", "% acumulada de filmes")
plt.title("Média de votos em filmes MovieLens 100K com pelo menos 10 votos")
```

Com esta visualização podemos tirar conclusões do tipo "Um filme com nota 3.9 está no topo 80% dos filmes" ou "Um filme com nota 2.5 está acima de nem 10% dos filmes." Mas, claro, é apenas uma forma visual, um rápido insight. Não se compara à uma analise estatistica detalhada.

# Analisando a distribuição dos dados capturados de outros campos do TMDB

```python
tmbd_com_mais_de_10_votos.head(1)
```

```python
ax = sns.displot(tmbd_com_mais_de_10_votos.vote_count, kde=True)
ax.set_axis_labels("Número de Votos", "Densidade")
plt.title("Número de votos em filmes no TMDB 5000 com 10 ou mais votos")
```

Poucos filmes são realmente muito populares, o que faz com que poucos tenham um alto número de votos.

```python

ax = sns.displot(tmdb.query("budget > 0").budget, kde=True)
ax.set_axis_labels("Budget (gastos)", "Densidade")
plt.title("Gastos em filmes no TMDB 5000")
```

Poucos filmes possuem alto orçamento, explicando o gráfico inclinado à esquerda.

```python
ax = sns.displot(tmdb.popularity, kde=True)
ax.set_axis_labels("Popularidade", "Densidade")
plt.title("Popularidade em filmes no TMDB 5000")
```

Poucos filmes possuem alta popularidade para o TMDB.

```python
ax = sns.displot(tmdb.query("runtime > 0").runtime.dropna(), kde=True)
ax.set_axis_labels("Duração", "Densidade")
plt.title("Duração de filmes no TMDB 5000")
```

```python
ax = sns.displot(tmdb.query("runtime > 0").runtime.dropna(), kind="ecdf")
ax.set_axis_labels("Duração", "% de filmes")
plt.title("Duração de filmes no TMDB 5000")
```

Para saber por exemplo, qual a duração mínima dos top 20% de filmes(em duração) no IMDB...

```python
tmdb.query("runtime>0").runtime.dropna().quantile(0.8)
```

Vemos que 80% dos filmes possuem 121 minutos ou menos de duração, só 20% dos filmes do dataset são mais longos que isso.
